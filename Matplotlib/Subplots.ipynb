```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,6,0.45)
y = 2 ** x
```

```python
a = np.arange(0,10,1.2)
b = 2 + a
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2, dpi = 200)

axes[0][0].plot(x,y)
axes[0][0].set_xlabel('Axes(0,0) - X axis')
axes[1][1].plot(a,b)
plt.tight_layout()
```
