```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Legends

```python
 x = np.arange(0,7,6)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x, 2**x, label = 'X vs 2^x')
plt.legend()
```

# Colors and Styles

```python
a = [1,2,3,4,5,6]
b = [2,4,6,8,10,12]
fig2 = plt.figure()
axes1 = fig2.add_axes([0,0,1,1])
axes1.plot(a, b, label = 'X vs 2^x', color = 'Maroon', ls = '--', marker = 'v')
plt.legend()
```
