```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,3,1)
y = 2 * x
```

```python
a = np.arange(3,7,1)
b = a ** 3
```

```python
fig = plt.figure(dpi = 200)

# Bigger Graph
axes1 = fig.add_axes([0,0,1,1])
axes1.set_title('X-Y Relation Graph')
axes1.set_xlabel('X Value')
axes1.set_ylabel('Y Value')

# Smaller Graph
axes2 = fig.add_axes([0.2,0.6,0.3,0.3])
axes2.set_title('A-B Relation Graph')
axes2.set_xlabel('A Value')
axes2.set_ylabel('B Value')

axes1.plot(x,y)
axes2.plot(a,b)

fig.savefig('figure-twoplots.jpg', bbox_inches = 'tight', dpi = 200)
```
