# Basics of Matplotlib

*Importing the matplotlib*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,2)
y = np.arange(5,15,2)
```

```python
x
```

```python
y
```

*Functional Parameters*

```python
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(0,10)
plt.ylim(0,20)
plt.title('Demo Graph')
plt.plot(x,y)
plt.savefig('firstdemoplot.jpg')
```
