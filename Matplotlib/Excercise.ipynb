# Excercise 1

*Plot world famous equation from Einstein **E = mc2**
where **m** is simply 11 evenly spaced values representing from 0 to 10 grams. **E** should be equivalent energy for the mass*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
m = np.linspace(0,10,11)
print(m)
```

```python
c =2.9979 * (10**8)
```

```python
E = m * (c**2)
print(E)
```

**Plotting Part :** 
*Recreate the plot which maps out E = mc2 using the arrays we created in the previous task. Note the labels, titles and colors and axis limits*

```python
fig = plt.figure(dpi = 200)
axes = fig.add_axes([0,0,1,1])
axes.plot(m, E, marker = '^', ls = '--')
axes.set_xlabel('Mass in Grams')
axes.set_ylabel('Energy in Joules')
axes.set_title('E = mc2')
```

# Excercise 2

**Yield Curve Data :**
*Data from US treasury Dept. The data shows interest paid for a US treasury bond for a certain contract length. The label list shows the correspoding contract length per index position*

```python
labels = ['1 Mo', '3 Mo', '6 Mo', '1 Yr', '2 Yr', '3 Yr', '5 Yr', '7 Yr', '10 Yr', '20 Yr', '30 Yr']
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
fig = plt.figure(dpi=200)
axes = fig.add_axes([0,0,1,1])
axes.plot(labels,july16_2007,label='July 16 2007', color = 'Red', ls = '--', marker = '^')
axes.plot(labels,july16_2020,label='July 16 2020', color = 'Blue', ls = '--', marker = '^')
plt.legend()
```

***Now try doing subplots with 2 rows and 1 column***

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, dpi = 200)
axes[0].plot(labels,july16_2007,label='July 16 2007', color = 'Red', ls = '--', marker = '^')
axes[1].plot(labels,july16_2020,label='July 16 2020', color = 'Blue', ls = '--', marker = '^')
plt.legend()
```
