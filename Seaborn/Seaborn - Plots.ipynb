# Distribution Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('SalesRecords.csv')
```

```python
df1.head(3)
```

***Rug Plot***

```python
plt.figure(dpi = 200, figsize = (7,4))
sns.rugplot(data=df1, x = 'Units Sold', height = 0.25)
```

***Histogram and Kernel Density Estimation***

```python
plt.figure(dpi =250, figsize = (9,5))
sns.histplot(df1, x = 'Units Sold', bins = 20, hue = 'Order Priority')
```

```python
plt.figure(dpi =400, figsize = (9,5))
sns.histplot(df1, x = 'Units Sold', bins = 20, hue = 'Order Priority', kde = True)
```

# Categorical Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfr = pd.read_csv('SalesRecords.csv')
```

```python
dfr.head(3)
```

```python
df['Order Priority'].value_counts()
```

```python
plt.figure(dpi = 275, figsize = (15,5))
sns.countplot(dfr, x = 'Item Type', hue ='Order Priority')
```

```python
plt.figure(dpi = 300, figsize = (15,5))
sns.barplot(dfr, x = 'Item Type', y = 'Total Cost', estimator = np.mean, errorbar = 'sd', hue = 'Order Priority')
```

```python
plt.figure(dpi = 300, figsize = (15,10))
sns.boxplot(dfr, y = 'Item Type', x = 'Total Cost')
```

```python
plt.figure(dpi = 300, figsize = (15,10))
sns.violinplot(dfr, y = 'Order Priority', x = 'Total Cost')
```

```python
plt.figure(dpi = 300, figsize = (15,5))
sns.swarmplot(dfr, x = 'Order Priority', y = 'Total Cost', hue = 'Sales Channel', size = 2)
```

# Grid Plots

```python
plt.figure(dpi =250, figsize = (10,5))
sns.catplot(dfa, x = 'gender', y = 'math score', kind = 'box', col = 'lunch')
```

```python
g = sns.PairGrid(dfa, hue = 'gender')
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
```
