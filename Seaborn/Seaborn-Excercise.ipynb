```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('application_record.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
emp = df[df['DAYS_EMPLOYED']<0] * -1
```

```python
emp ['DAYS_BIRTH'] = emp ['DAYS_BIRTH'] * -1
```

```python
plt.figure(dpi = 250, figsize= (10,5))
sns.scatterplot(emp, x = 'DAYS_BIRTH', y = 'DAYS_EMPLOYED', alpha = 0.01, lw = 0)
```
